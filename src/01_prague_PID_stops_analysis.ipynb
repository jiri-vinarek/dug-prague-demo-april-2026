# Analysis of Prague PID Dataset

Let's explore https://pid.cz/en/opendata/ dataset - in particular, the PID stop list (http://data.pid.cz/stops/json/stops.json)

## Data Loading
Fetch the stop list JSON from the PID open data API and flatten it into a pandas DataFrame.

```python
import requests
import json

response = requests.get("http://data.pid.cz/stops/json/stops.json")
response
```

```python
data = response.json()

# Show top-level keys and metadata
print(f"Top-level keys: {list(data.keys())}")
print(f"Generated at: {data['generatedAt']}")
print(f"Data format version: {data['dataFormatVersion']}")
print(f"Number of stopGroups: {len(data['stopGroups'])}")
print(f"\nFirst stopGroup:")
print(json.dumps(data['stopGroups'][0], indent=2, ensure_ascii=False))
```

## Visualizations
Geographic distribution and transport type breakdown of PID stop groups.

```python
import pandas as pd

df = pd.json_normalize(data["stopGroups"])
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
display(df.head(20))
```

```python
import pandas as pd
import plotly.express as px

fig = px.scatter_mapbox(
    df,
    lat="avgLat",
    lon="avgLon",
    hover_name="name",
    zoom=10,
    height=600
)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
fig = px.density_mapbox(
    df,
    lat="avgLat",
    lon="avgLon",
    hover_name="name",
    radius=6,
    zoom=7,
    height=600,
    title="PID Stop Density Heatmap"
)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 40, "l": 0, "b": 0})
fig.show()
```

```python
type_counts = df["mainTrafficType"].value_counts().reset_index()
type_counts.columns = ["transport_type", "stop_count"]

fig = px.bar(
    type_counts,
    x="transport_type",
    y="stop_count",
    color="transport_type",
    title="Number of Stop Groups by Transport Type",
    labels={"transport_type": "Transport Type", "stop_count": "Stop Groups"},
    text="stop_count"
)
fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.scatter_mapbox(
    df,
    lat="avgLat",
    lon="avgLon",
    color="mainTrafficType",
    hover_name="name",
    hover_data=["municipality", "mainTrafficType"],
    zoom=10,
    height=700,
    title="PID Stops by Transport Type"
)
fig.update_layout(
    mapbox_style="carto-positron",
    margin={"r": 0, "t": 40, "l": 0, "b": 0}
)
fig.show()
```

## Checking stops

Let's explode stops list and create `df_stops` dataframe.

```python
df_stops = df.explode("stops").reset_index(drop=True)
stops_normalized = pd.json_normalize(df_stops["stops"])
stops_normalized = stops_normalized.rename(columns={"mainTrafficType": "stopTrafficType"})
df_stops = pd.concat([df_stops.drop(columns=["stops"]).reset_index(drop=True), stops_normalized], axis=1)

print(f"Shape: {df_stops.shape}")
print(f"Columns: {list(df_stops.columns)}")
display(df_stops.head(10))
```

```python
fig = px.scatter_mapbox(
    df_stops,
    lat="lat",
    lon="lon",
    color="zone",
    hover_name="altIdosName",
    hover_data=["id", "platform", "zone", "mainTrafficType"],
    zoom=10,
    height=700,
    title="PID Stops by Zone"
)
fig.update_layout(
    mapbox_style="carto-positron",
    margin={"r": 0, "t": 40, "l": 0, "b": 0}
)
fig.show()
```
