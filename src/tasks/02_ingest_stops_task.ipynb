# 02 – Ingest Stops

Reads the raw PID stops JSON file from the volume and materialises it into a bronze Delta table, adding the source file name as a metadata column.

## Inspect Input Parameters

```sql
SELECT
  :raw_stops_volume AS raw_stops_volume,
  :bronze_stops_table AS bronze_stops_table
```

## Ingest Data

> **NOTE:**  The query below loads all files present in `raw_stops_volume` on each run. This approach ingests the entire dataset every time, which may impact performance as the data grows. If this becomes a performance issue, consider using Spark Declarative Pipelines to load data incrementally.

```sql
CREATE OR REPLACE TABLE IDENTIFIER(:bronze_stops_table) AS
SELECT
  *,
  regexp_extract(_metadata.file_path, '[^/]+$', 0) AS file_name
FROM read_files(:raw_stops_volume, format => 'json', multiLine => true)
```

```sql
SELECT COUNT(*) FROM IDENTIFIER(:bronze_stops_table)
```
