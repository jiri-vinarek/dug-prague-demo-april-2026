# 01 – Fetch Stops

Downloads the current PID (Prague Integrated Transport) stop list from the public API and persists the raw response as a file in the configured Unity Catalog volume.

```python
import os
import sys
sys.path.insert(0, '../exports')
from fetching_functions_export import *
```

## Inspect Input Parameters

```python
raw_stops_volume = dbutils.widgets.get("raw_stops_volume")
print(f"raw_stops_volume = {raw_stops_volume}")
```

## Fetch Data

```python
data = fetch_pid_stops()
stops_filename = get_stops_file_name()

output_path = os.path.join(raw_stops_volume, stops_filename)
with open(output_path, "wb") as f:
    f.write(data)

print(f"Written {len(data):,} bytes to {output_path}")
```
