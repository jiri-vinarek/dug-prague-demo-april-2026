```python
# | export
import requests
import time
```

```python
# | export

def fetch_pid_stops() -> bytes:
    """Fetch PID stops data from the API and return raw response content.
    
    Raises:
        requests.HTTPError: If the API returns a non-2xx status.
    """
    PID_STOPS_URL = "http://data.pid.cz/stops/json/stops.json"
    response = requests.get(PID_STOPS_URL, timeout=30)
    response.raise_for_status()
    return response.content
```

```python
from fastcore.test import *
from operator import gt
import nbdev
```

```python
# Returns bytes
result = fetch_pid_stops()
test_eq(type(result), bytes)

# Non-empty response
test(len(result), 0, gt)

print("All tests passed!")
```

```python
# | export

def get_stops_file_name() -> str:
    """Generate a unique file name for stops data using the current unix timestamp.

    Returns:
        A string in the format `<unix_timestamp>.json`.
    """
    unix_timestamp = int(time.time())
    return f"{unix_timestamp}.json"
```

```python
result = get_stops_file_name()
test_eq(result.endswith('.json'), True)
test(int(result.removesuffix('.json')), 0, gt)

print(f"Got: {result}")
```
